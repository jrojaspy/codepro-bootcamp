```python
## VARIABLE GLOBAL

intentos = 0

def conteo_de_intentos(attemps):

    while attemps < 5:
        #### FUNCION REVISORA
        attemps = attemps + 1
        print(f"Le quedan {6 - attemps} intentos")

conteo_de_intentos(intentos)
```

```python
### FUNCION DE GRILLA

letras_verificadas = ['[[p]]', 'a', 't', '((o))', 's']
letras_coloreadas = []

def colorear_letras(letras_verificadas):

    for letra in letras_verificadas:

        if "[[" in letra:
            print("\033[32m" + letra + "\33[0m")
            letras_coloreadas.append("\033[32m" + letra + "\33[0m")
        elif "((" in letra:
            print("\033[33m" + letra + "\33[0m")
            letras_coloreadas.append("\033[33m" + letra + "\33[0m")
        else:
            print("\033[31m" + letra + "\33[0m")
            letras_coloreadas.append("\033[31m" + letra + "\33[0m")
    
    return letras_coloreadas

print(*colorear_letras((letras_verificadas)))
```

```python
palabra_usuario = input("Escriba su palabra:")
palabra_secreta = "perro"

letras_verificadas = []
cantidad_letras = 5


##### FUNCION VERIFICADORA

def verificar_palabra(palabra_usuario, palabra_secreta):

    for letra in range(cantidad_letras):
        
        ### VARIABLES LOCALES

        palabras_iguales = palabra_usuario[letra] == palabra_secreta[letra]
        letra_en_palabra_secreta = palabra_usuario[letra] in palabra_secreta

        if palabras_iguales:
            letras_verificadas.append("\033[32m" + palabra_usuario[letra] + "\33[0m")
        elif letra_en_palabra_secreta:
            letras_verificadas.append("\033[33m" + palabra_usuario[letra] + "\33[0m")
        else:
            letras_verificadas.append("\033[31m" + palabra_usuario[letra] + "\33[0m")

    return letras_verificadas

(verificar_palabra(palabra_usuario,palabra_secreta))
print(*letras_verificadas)
```
