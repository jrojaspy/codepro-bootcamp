```python
# Crear una función que reciba dos parámetros (una palabra y un número), y que retorne True si el número es igual a la cantidad de letras de la palabra. En caso contrario, que retorne False.
# Pista: cantidad de letras de una palabra en python

letter = input("Ingrese la palabra: ")
number = int(input("Ingrese la palabra: "))

def cantidad_de_caracteres(letter, number):
    if len(number) == len(letter):
        return True
    else:
        return False

# return permite generar datos que se pueden almacenar en variables o estructuras de datos y utilizarlos en el programa.

caracteres = cantidad_de_caracteres(letter, number)

print(caracteres)
print(type(caracteres))
```

```python
# Crear función que revise si una palabra contiene una letra. 
# En caso de que la contenga, retornar la letra entre corchetes []. De lo contrario, retornar solo la letra.

palabra = input("Ingresar palabra: ")
letra_a_verificar = input("Ingresar letra: ")

def revisar_palabra(word, character):
    if character in word:
        return f"[{character}]"
    else:
        return character

# no se puede usar for porque corta la iteracion 

revisar_palabra(palabra, letra_a_verificar)
```

```python
# Pedir al usuario que ingrese 1 palabra, luego convertir esa palabra a una lista con sus caracteres con list(). 
# Agregar esa lista de caracteres a una lista llamada palabras_ingresadas. Repetir el proceso 3 veces. 
# Cada vez que se agregue una lista de caracteres, imprimir la lista palabras_ingresadas.

palabra = ""
palabras_ingresadas = []

for palabra in range(3):
    palabra = list(input("Ingrese su palabra: "))
    palabras_ingresadas.append(palabra)
    print(palabras_ingresadas)

# print(palabras_ingresadas)
```

```python
# Dada la lista de ejemplo: 

# lista = [['h', 'o', 'l', 'i', 's'], ['h', 'e', 'l', 'l', 'o'], ['h', 'a', 'l', 'l', 'o']]

# Crear una función que imprima una grilla (lista de listas), una línea por cada lista dentro de la lista principal

lista_para_grilla = [['h', 'o', 'l', 'i', 's'], ['h', 'e', 'l', 'l', 'o'], ['h', 'a', 'l', 'l', 'o']]

def imprimir_grilla(lista):
    for elementos in lista:
        print(elementos)

imprimir_grilla(lista_para_grilla)
```
