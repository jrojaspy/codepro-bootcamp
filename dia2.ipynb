```python
# Escribe un programa que pregunte al usuario por su edad y determine si tiene edad suficiente para comprar y consumir alcohol.
# Si el usuario tiene mas de 18 pero menos de 21 impirmir "Podes tomar pero no comprar"
# Si el usuario tiene menos de 18 imprimir "Juguito nomas"
# Si el usuario tiene mas de 21 imprimir "Hace lo que quieras"

# utilizando los pilares de pensamiento
# descomponer, reconocimiento de patron, abstraccion y algoritmo

edad_usuario = int(input("Cual es su edad?: "))

if edad_usuario < 18:
    print("Juguito nomas")
elif edad_usuario <= 21:
    if edad_usuario >= 18:
        print("Podes tomar pero no comprar")
    else:
        print("Podes tomar pero no comprar")
else:
    print("Hace lo que quieras")
```

```python
# Escribir un programa en el que se pregunte al usuario por una temperatura (en grados Celsius), imprima la temperatura escrita, y luego clasificar la temperatura ingresada en los siguientes niveles:

# Entre -10 y 18 grados, que imprima "Fresquete".
# Entre 19 y 30 grados, que imprima "Calor'i".
# Entre 31 y 45 grados, que imprima "Hakuuuuuu".
# Para todo lo demás, imprimir "Ñamano mba'e!".

temperatura = int(input("Cual es la temperatura registrada en Grados Celsius?: "))
print(f"La temperatura registrada es de: {temperatura}")

def info_de_temperatura(temperatura):

    if temperatura >= -10 and temperatura <= 18:
        print("Fresquete")

    elif temperatura >= 19 and temperatura <= 30:
        print("Calor-i")

    elif temperatura >= 31  and temperatura <= 45:
        print("Hakuuuuuuuu")

    else:
        print("Ñamano mba'e!")

info_de_temperatura(temperatura)
```

```python
# Escribir un programa que almacene una contraseña en una variable, y pregunte al usuario por la contraseña hasta que introduzca la contraseña correcta.

# Cuando el usuario ingrese la contrasena correcta, imprimir un mensaje secreto.

contrasenha_base = "Penguin"

while True:
    contrasenha_usuario = input("Introduzca su contrasenha, por favor")
    if contrasenha_base != contrasenha_usuario:
        print("Contrasenha incorrecta, vuelva a intentarlo")
    else:
        print("Contrasenha correcta")
        print("Mensaje secreto: Pelea por tus suenhos")
        break
```

```python
# Crear una lista de 5 elementos. Luego, seguir los siguientes pasos.

# Agregar un elemento al final de la lista.
# Modificar el segundo elemento.
# Eliminar el tercer elemento.

lista_frutas = ["naranja", "sandia", "pera", "banana", "pomelo"]
print(lista_frutas)

# agregar un elemento al final
lista_frutas.append("kiwi")
print(lista_frutas)

# modifiicar el segundo elemento
lista_frutas[1] = "limon"
print(lista_frutas)

# eliminar tercer elemento
lista_frutas.pop(2)
print(lista_frutas)
```

```python
# Escribir un programa que itere en un rango de 10 veces. Por cada iteracion multiplicar el iterador por dos e imprimir el resultado.

# for iterator in range(10):
#     print(iterator * 2)

rango = 10
lista_impresion = []

for iterator in range(rango):
    lista_impresion.append(iterator*2)
    for i in lista_impresion:
        print(i)
```

```python
# Recorrer los elementos de una lista de marcas e imprimir cada marca solamente si tiene 6 letras o mas>

lista_marcas = ["puma", "adidas", "nike", "reebok", "topper", "kirios", "new balance"]

for marca in (lista_marcas):
    if len(marca) >= 6:
        print(marca)
```

```python
# Solicitar al usuario que ingrese una frase.

# Imprimir un mensaje cada vez que se encuentre una letra "o".

palabra = input("Ingrese su palabra para analizar: ")
counter = 0
print(f"La palabra ingresada es {palabra}")

for letra in palabra:
    
    if letra == "o" or letra == "O" :
        counter = counter + 1
        print("Es una  letra 'O'")  

print(f"Usaste la letra 'O', {counter} veces")
```
